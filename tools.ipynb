```python
import subprocess

source_template = "/home/frozzzen/Documents/Github/layered-mesh-gaussian/data/dancer/meshes_TSMC/original/dancer_{:03d}.obj"
output_template = "/home/frozzzen/Documents/Github/layered-mesh-gaussian/data/dancer/meshes_TSMC/dancer_{:04d}.obj"

target_mesh = "/home/frozzzen/Documents/Github/layered-mesh-gaussian/data/hotdog/mesh.obj"
target_mesh = "/home/frozzzen/Documents/Github/layered-mesh-gaussian/data/dancer/meshes_no_texture/dancer_0001.obj"

#scale = "0.0015"

for i in range(1, 11):  # 0001 to 0010
    source = source_template.format(i-1)
    output = output_template.format(i)
    print(f"Processing frame {i}...")
    print(f"Source: {source}")
    print(f"Output: {output}")
    cmd = [
        "python", "align_mesh.py",
        "--source", source,
        "--target", target_mesh,
        "--output", output,
        "--auto_scale",
        #"--scale", scale,
        "--anchor", "center",
        "--bbox_type", "aabb"
    ]

    print("Running:", " ".join(cmd))
    subprocess.run(cmd, check=True)

print("All frames processed.")
```
